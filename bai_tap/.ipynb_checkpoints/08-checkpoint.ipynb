```python
import matplotlib.pyplot as plt
```

```python
# Vẽ biểu đồ: Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

# Vẽ đường 1
plt.plot(x, y, 'bo-', label='Dữ liệu') 
# Vẽ đường 2
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label='Dữ liệu-1')
# Hiển thị chú giải
plt.legend()
# Đặt nhãn cho trục x
plt.xlabel('Trục x')
# Đặt nhãn cho trục y
plt.ylabel('Trục y')
# Hiển thị biểu đồ
plt.show()
```

```python
# Vẽ biểu đồ cột Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

# Vẽ biểu đồ cột
plt.bar(categories, values, color='g', alpha=0.6)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
# Vẽ biểu đồ hình tròn Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

# Vẽ biểu đồ hình tròn
plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

# Vẽ biểu đồ phân tán
plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')
plt.title('Biểu đồ phân tán')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
# Hiển thị chú giải
plt.legend()
plt.show()
```

```python
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

# Vẽ biểu đồ hộp
plt.boxplot(data)  # Vẽ biểu đồ hộp
plt.title('Biểu đồ hộp')
plt.ylabel('Giá trị')
plt.show()
```

```python
# Vẽ biểu đồ violin
plt.violinplot(data)
plt.title('Biểu đồ violin')
plt.ylabel('Giá trị')
plt.show()
```
