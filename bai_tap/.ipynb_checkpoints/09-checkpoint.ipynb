```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
dir = 'C:/Users/admin/Desktop/nhung_mon_hoc/thi_giac_may_tinh/anh'
image_path = dir + '/anh.png'
```

```python
# doc anh
im = Image.open(image_path)
plt.imshow(im)
```

```python
# chuyen sang anh sang
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# tao mot hinh ve moi
plt.figure()
plt.hist(im_array.flatten(), bins = 128)
plt.title('Bieu do Histogram cua anh')
plt.xlabel('Gia tri cua pixel')
plt.ylabel('So luong')
plt.show()
```

```python
# lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(im_gray, origin = "image")
plt.show()
```

```python
# doc anh va hien thi anh voi truc toa do
im = Image.open(image_path).convert("L")

# hien thi anh tren truc toa do
plt.imshow(im)

# tao mot hinh ve moi
plt.figure()

# cho mau la xam
plt.gray()

# ve bieu do duong bien cua anh
plt.contour(im, origin = "image")

# dam bao ty le truc x va truc y tren bieu do la bang nhau
plt.axis('equal')
plt.show()
```
